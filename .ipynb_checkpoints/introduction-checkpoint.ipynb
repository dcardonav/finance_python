## Risk-Adjusted Returns

```python
df_returns.std()

# calculating volatility step by step
deviations = df_returns - df_returns.mean()
sq_dev = deviations ** 2
vari = sq_dev.mean()
volatility = np.sqrt(vari) # remember that the standard deviation divides by n-1

print(volatility)
print(df_returns.std())

vari = sq_dev.sum()/(df_returns.shape[0] -1)
volatility = np.power(vari, 0.5)
print(volatility)

ann_volat = df_returns.std()*np.sqrt(12) # annualizing volatility
print(ann_volat)
```

```python
# loading the dataset, indicating if it has header and telling it to try to figure out dates and specifying a value
# for the na's (99.99)
returns = pd.read_csv('course_1/Portfolios_Formed_on_ME_monthly_EW.csv', header=0, index_col=0,parse_dates=True,
                      na_values=99.99)

columns = ['Lo 10', 'Hi 10']
returns = returns[columns]/100
returns.columns = ['SmallCap', 'LargeCap']
returns.plot.line()

returns.std()

annualized_vol = returns.std()*np.sqrt(12)
annualized_vol

# annualizing returns
return_per_month = (returns + 1).prod() ** (1/returns.shape[0]) -1
return_per_month

annualized_return = (return_per_month + 1) ** (12) - 1
annualized_return

# risk-adjusted return
print(annualized_return / annualized_vol)

riskfree_rate = 0.03
excess_return = annualized_return - riskfree_rate
sharpe_ratio = excess_return / annualized_vol
sharpe_ratio # SmallCap give you a slightly superior return over LargeCap
```

## Risk-Adjusted Returns (Max Drawdown)

A worst-case measure: if you buy at the peak-price and sell at the lowest.

```python


returns.index = pd.to_datetime(returns.index, format='%Y%m')
returns.index = returns.index.to_period('M')
```

```python
# code to compute drawdowns
# 1. Compute a wealth index
# 2. Compute previous peaks
# 3. Compute the drawdown

wealth_index = 1000*(1+returns.LargeCap).cumprod() # multiply by each monthly-return
wealth_index.plot.line()


previous_peaks = wealth_index.cummax() # maximum up to a certain point
previous_peaks.plot.line()

drawdown = (wealth_index - previous_peaks)/ previous_peaks # maximum loss
plt.figure()
drawdown.plot()

drawdown['1975':].idxmin() # index of the minimum position
```

```python
drawdown(returns.LargeCap)[['wealth', 'peaks']].plot()
drawdown(returns[:'1950'].LargeCap)[['wealth', 'peaks']].plot()
drawdown(returns.LargeCap).min()
```
